### Train a Spam Detector Algorithm

You are working for an email service provider and have been tasked with training an algorithm that recognizes whether an email is spam or not from a given dataset and checking its performance.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
csv_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter01/Dataset/dataset_44_spambase.csv'
```

```python
csv_df = pd.read_csv(csv_url)
```

```python
csv_df.head()
```

```python
target = csv_df.pop('class')
```

```python
target.head()
```

```python
csv_df.head()
```

```python
rf_model = RandomForestClassifier(random_state=1)

rf_model.fit(csv_df, target)
```

```python
preds = rf_model.predict(csv_df)
preds
```

```python
accuracy_score(target, preds)
```
